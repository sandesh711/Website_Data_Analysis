##### csv file is from mysql database visitor table results as per query. Kept small date window since data was large.  
SELECT  
SUBSTRING_INDEX(SUBSTRING_INDEX(url, '/tag/', -1), '/', 1) AS tag  
from visitor  
WHERE   
url like "%/tag/%" and referer not like '%sandesh711.com%'  
and date(time)>'2025-08-01'  
results then exported to CSV  

```python
import pandas as pd
df=pd.read_csv(r'C:\Users\sande\Desktop\sandesh\coursera\IBM_DS\visitor.csv')
# to remove junk values
df=df[~df['tag'].astype(str).str.startswith(('%','Ã','à²'))] 
df=df[df['tag'].astype(str)!='webfonts']
# converting df value to string
tag_string = ' '.join(df['tag'].astype(str))
tag_string
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud(width=800, height=500, background_color='white').generate(tag_string)
# Plot the word cloud
plt.figure(figsize=(12, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # No axes
plt.show()
```
